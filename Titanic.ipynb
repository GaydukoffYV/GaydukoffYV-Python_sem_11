```python
# Очистка от нулевых значений
df.dropna(subset=['Age', 'Cabin', 'Embarked'])
```

```python
df_clear = df.dropna(subset=['Age', 'Cabin', 'Embarked'])
```

```python
# Данные по именам и транспортным расходам
df_clear[['Name', 'Fare']]
```

```python
# Данные по мужчмнам младше 35 лет
df_clear[(df['Age'] < 35) & (df['Sex'] == 'male')]
```

```python
# Определим максимальный, минимальный и средний возраст пассажиров
print(df_clear['Age'].max())
print(df_clear['Age'].min())
print(df_clear['Age'].mean())
```

```python
# Определим максимальные транспортные расходы пассажиров в возрасте старше 40 лет
df_clear.loc[(df['Age'] > 40),['Fare']].max()
```

```python
# Узнаем в какой каюте находится самый возрастной пассажир
df_clear.loc[(df['Age'] == (df['Age']).max()), ['Cabin']].max()
```
